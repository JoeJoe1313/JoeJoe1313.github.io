```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()  # Directed graph for hierarchical visualization

# Add edges (the nodes are inferred from edges)
G.add_edges_from([
    (1, 2), (1, 3), (1, 4), (1, 5), (1, 6), (1, 7), (1, 8), (1, 9),
    (4, 10), (4, 11), (4, 12), (4, 13), (4, 14), (4, 15), (4, 16), (4, 17),
])

# Use `shell_layout` to make it top-down
pos = nx.shell_layout(G)  

# Draw graph
plt.figure(figsize=(6, 6))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15, arrows=True)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout

# Create a directed graph (DiGraph) for hierarchy
G = nx.DiGraph()

# Add edges to represent a tree structure
G.add_edges_from([
    ("root", "FTL"), ("root", "FTR"), ("root", "FBL"), ("root", "FBR"),
    ("root", "BTL"), ("root", "BTR"), ("root", "BBL"), ("root", "BBR"),
    ("FBR", "X1"), ("FBR", "X2"), ("FBR", "X3"), ("FBR", "X4"),
    ("FBR", "X5"), ("FBR", "X6"), ("FBR", "X7"), ("FBR", "X8")
])

# Use Graphviz layout (dot) to make it top-down
pos = graphviz_layout(G, prog="dot")

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="darkblue", 
        node_size=2000, font_size=10, arrows=True)

# plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_agraph import graphviz_layout

# Create a directed graph (DiGraph) for hierarchy
G = nx.DiGraph()

# Add edges to represent a tree structure
G.add_edges_from([
    ("root", "UR_0"), ("root", "UL_0"), ("root", "LL_0"), ("root", "LR_0"),
    ("UR_0", "UR"), ("UR_0", "UL"), ("UR_0", "LL"), ("UR_0", "LR"),
    ("UL_0", "UR1"), ("UL_0", "UL1"), ("UL_0", "LL1"), ("UL_0", "LR1"),
    ("LL_0", "UR2"), ("LL_0", "UL2"), ("LL_0", "LL2"), ("LL_0", "LR2"),
    ("LR_0", "UR3"), ("LR_0", "UL3"), ("LR_0", "LL3"), ("LR_0", "LR3"),
])

# Use Graphviz layout (dot) to make it top-down
pos = graphviz_layout(G, prog="dot")

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="darkblue", 
        node_size=500, font_size=10, arrows=True)

# plt.show()
```
